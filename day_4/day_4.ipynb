## 1. Puzzle 1

### 1.1 My Solution

```python
import numpy as np
```

```python
def check_winner(array):
    for row in array:
        if np.count_nonzero(row == 'X') == 5:
            return True
            break
        else:
            pass
```

#### Using the example

```python
example = ['7,4,9,5,11,17,23,2,0,14,21,24,10,16,13,6,15,25,12,22,18,20,8,19,3,26,1\n',
 '\n',
 '22 13 17 11  0\n',
 ' 8  2 23  4 24\n',
 '21  9 14 16  7\n',
 ' 6 10  3 18  5\n',
 ' 1 12 20 15 19\n',
 '\n',
 ' 3 15  0  2 22\n',
 ' 9 18 13 17  5\n',
 '19  8  7 25 23\n',
 '20 11 10 24  4\n',
 '14 21 16 12  6\n',
 '\n',
 '14 21 17 24  4\n',
 '10 16 15  9 19\n',
 '18  8 23 26 20\n',
 '22 11 13  6  5\n',
 ' 2  0 12  3  7\n']

draw_sequence = example[0].strip().split(',')
boards = [row for row in example[2:] if row != '\n'] 

print("Total number of boards:", int(len(boards)/5))
boards_as_3d = np.array([x for row in boards for x in row.strip().split()]).reshape(int(len(boards)/5),5,5)

for idx, chosen_num in enumerate(draw_sequence):
    print(f"Round {idx}: The chosen number is {chosen_num}.")
    
    # Mark the chosen number in the boards with an 'X'
    boards_as_3d = np.where(boards_as_3d == chosen_num, 'X', boards_as_3d)
    
    # Let's check whether there are any winners
    for board in boards_as_3d:
        
        # Check each 5x5 board to see if any board wins
        if check_winner(board) or check_winner(board.T):
            print("We have a winner!!\nThe winning board is:\n", board)
            break
            
        else:
            pass
        
    else:
        continue # only executed if the inner loop did NOT break
        
    break # only executed if the inner loop DID break

unmarked_nums = [int(ele) for row in board.tolist() for ele in row if ele != 'X'] 
print(f"The last number that was called before there's a winner: {chosen_num}")
print(f"Sum of unmarked numbers: {sum(unmarked_nums)}")
print(f"Final score: {sum(unmarked_nums) * int(chosen_num)}")
```

## 2. Puzzle 2

### 2.1 My Solution

```python
with open('data/input.txt') as f:
    lines = f.readlines()

draw_sequence = lines[0].strip().split(',')
boards = [row for row in lines[2:] if row != '\n'] 

print("Total number of boards:", int(len(boards)/5))
boards_as_3d = np.array([x for row in boards for x in row.strip().split()]).reshape(int(len(boards)/5),5,5)
# winner_list = []

for idx, chosen_num in enumerate(draw_sequence):
    print(f"\nRound {idx}: The chosen number is {chosen_num}.")
    
    # Mark the chosen number in the boards with an 'X'
    boards_as_3d = np.where(boards_as_3d == chosen_num, 'X', boards_as_3d)

    # Initiate a list to collect indices of winning boards
    winner_list = []
    
    # Let's check whether there are any winners
    for board_idx, board in enumerate(boards_as_3d):
        
        # Check each 5x5 board to see if any board wins
        if check_winner(board) or check_winner(board.T):
            print("We have a winner!!\nThe winning board is:\n", board)
            winner_list.append(board_idx)
            
        else:
            pass
    
    # Remove the winning board from the array of boards
    boards_as_3d = np.array([board for i, board in enumerate(boards_as_3d.tolist()) if i not in winner_list])
    print("How many boards now:", len(boards_as_3d))

    # Break the loop at the last winner
    if len(boards_as_3d) == 0:
        break

unmarked_nums = [int(ele) for row in board.tolist() for ele in row if ele != 'X'] 
print(f"The last number that was called before there's a winner: {chosen_num}")
print(f"Sum of unmarked numbers: {sum(unmarked_nums)}")
print(f"Final score: {sum(unmarked_nums) * int(chosen_num)}")
```

```python
def find_count_of_easy_numbers(data):
    
    output_values = [entry.split('| ')[1].strip() for entry in data]
    num_segments_in_easy_numbers = [2, 4, 3, 7]
    return len([digit for value in output_values for digit in value.split() if len(digit) in num_segments_in_easy_numbers])

if __name__ == "__main__":
    
    # Example data
    example_data = [
        'be cfbegad cbdgef fgaecd cgeb fdcge agebfd fecdb fabcd edb | fdgacbe cefdb cefbgd gcbe\n', 
        'edbfga begcd cbg gc gcadebf fbgde acbgfd abcde gfcbed gfec | fcgedb cgb dgebacf gc\n', 
        'fgaebd cg bdaec gdafb agbcfd gdcbef bgcad gfac gcb cdgabef | cg cg fdcagb cbg\n',
        'fbegcd cbd adcefb dageb afcb bc aefdc ecdab fgdeca fcdbega | efabcd cedba gadfec cb\n',
        'aecbfdg fbg gf bafeg dbefa fcge gcbea fcaegb dgceab fcbdga | gecf egdcabf bgf bfgea\n',
        'fgeab ca afcebg bdacfeg cfaedg gcfdb baec bfadeg bafgc acf | gebdcfa ecba ca fadegcb\n',
        'dbcfg fgd bdegcaf fgec aegbdf ecdfab fbedc dacgb gdcebf gf | cefg dcbef fcge gbcadfe\n',
        'bdfegc cbegaf gecbf dfcage bdacg ed bedf ced adcbefg gebcd | ed bcgafe cdgba cbgef\n',
        'egadfb cdbfeg cegd fecab cgb gbdefca cg fgcdab egfdb bfceg | gbdfcae bgc cg cgb\n',
        'gcafb gcf dcaebfg ecagb gf abcdeg gaef cafbge fdbac fegbdc | fgae cfgab fg bagce\n'    
    ]
    
    assert find_count_of_easy_numbers(example_data) == 26
    
    # Read input file
    with open('data/input.txt') as f:
        data = f.readlines()

    print(f"Number of times 1, 4, 7 and 8 appear: {find_count_of_easy_numbers(data)}")
```

## 2. Puzzle 2

### 2.1 My Solution

```python
def get_easy_numbers(list_of_signals):
    """Given a list of signals, return signals corresponding to the digits 1, 4, 7 and 8."""
    
    for item in list_of_signals:
        if len(item) == 2:
            one = item
        elif len(item) == 4:
            four = item
        elif len(item) == 3:
            seven = item
        elif len(item) == 7:
            eight = item

    return one, four, seven, eight

def get_three(list_of_signals, signal_for_seven):
    """Given a list of signals and the signal corresponding to seven, return the signal corresponding to the digit 3."""
    
    for signal in list_of_signals:
        if (len(signal) == 5) & (set(signal_for_seven).intersection(set(signal)) == set(signal_for_seven)):
            return signal
            break
        else:
            pass
        
def get_nine(list_of_signals, signal_for_four):
    """Given a list of signals and the signal corresponding to four, return the signal corresponding to the digit 9."""

    for signal in list_of_signals:
        if (len(signal) == 6) & (set(signal_for_four).intersection(set(signal)) == set(signal_for_four)):
            return signal
            break
        else:
            pass
        
def get_zero(list_of_signals, signal_for_seven):
    """Given a list of signals and the signal corresponding to seven, return the signal corresponding to the digit 0."""

    for signal in list_of_signals:
        if (len(signal) == 6) & (set(signal_for_seven).intersection(set(signal)) == set(signal_for_seven)):
            return signal
            break
        else:
            pass
        
def get_five(list_of_signals, signal_for_six):
    """Given a list of signals and the signal corresponding to six, return the signal corresponding to the digit 5."""

    for signal in list_of_signals:
        if (len(signal) == 5) & (set(signal).intersection(set(signal_for_six)) == set(signal)):
            return signal
            break
        else:
            pass
        
def get_mapping(list_of_signals):
    """Given a list of signals, return the mapping of each signal to each digits as a dictionary.
    
    For example, the sample input ['be', 'cfbegad', 'cbdgef', 'fgaecd', 'cgeb', 'fdcge', 'agebfd', 'fecdb', 'fabcd', 'edb'] 
    will generate the following mapping:
    
    {'agebfd': 0, 'be': 1, 'fabcd': 2, 'fecdb': 3, 'cgeb': 4, 'fdcge': 5, 'fgaecd': 6, 'edb': 7, 'cfbegad': 8, 'cbdgef': 9}
    """
    
    # Get signals for easy numbers first, i.e., 1, 4, 7, 8
    one, four, seven, eight = get_easy_numbers(list_of_signals)
    mapping = {one: '1', four: '4', seven: '7', eight: '8'}
    for k, v in mapping.items():
        list_of_signals = [v if signal == k else signal for signal in list_of_signals]

    # Get signal for not-so-easy numbers sequentially 
    three = get_three(list_of_signals, seven)
    list_of_signals = ['3' if signal == three else signal for signal in list_of_signals]

    nine = get_nine(list_of_signals, four)
    list_of_signals = ['9' if signal == nine else signal for signal in list_of_signals]

    zero = get_zero(list_of_signals, seven)
    list_of_signals = ['0' if signal == zero else signal for signal in list_of_signals]

    six = [signal for signal in list_of_signals if len(signal) == 6][0]
    list_of_signals = ['6' if signal == six else signal for signal in list_of_signals]

    five = get_five(list_of_signals, six)
    list_of_signals = ['5' if signal == five else signal for signal in list_of_signals]

    two = [signal for signal in list_of_signals if len(signal) == 5][0]
    list_of_signals = ['2' if signal == two else signal for signal in list_of_signals]

    # Save each mapping of signal to number as a dictionary
    mapping = dict()
    mapping[zero], mapping[one], mapping[two] = 0, 1, 2
    mapping[three], mapping[four], mapping[five] = 3, 4, 5
    mapping[six], mapping[seven], mapping[eight], mapping[nine] = 6, 7, 8, 9

    return mapping        

def get_output_value(mapping, output_values):
    """Given a mapping and a list of signals representing the output values, return the output value in integer form."""
    output = ''
    for value in output_values:
        for k, v in mapping.items():
            if set(value) == set(k):
                output += str(v)

    return int(output)        

def get_sum_of_output_values(list_of_signals, output_values):
    """Given a list of list of signals, return the total sum of all output values."""
    output_value = 0
    for idx, signal in enumerate(list_of_signals):
        mapping = get_mapping(signal)
        output_value += get_output_value(mapping, output_values[idx])
    return output_value


if __name__ == '__main__':
    
    # Example 1
    example_1 = [
        'acedgfb cdfbe gcdfa fbcad dab cefabd cdfgeb eafb cagedb ab | cdfeb fcadb cdfeb cdbaf\n'
    ]
    signals_and_output_1 = [line.strip().split(" | ") for line in example_1][0]
    signals_1 = [signals_and_output_1[0].split()]
    output_1 = [signals_and_output_1[1].split()]
    
    mapping_1 = get_mapping(signals_1[0])
    output_value_1 = get_output_value(mapping_1, output_1[0])
    
    assert mapping_1['eafb'] == 4
    assert mapping_1['acedgfb'] == 8
    assert mapping_1['cdfbe'] == 5
    assert mapping_1['cefabd'] == 9
    assert mapping_1['cagedb'] == 0
    assert mapping_1['cdfgeb'] == 6
    assert output_value_1 == 5353

    # Example 2
    example_2 = [
        'be cfbegad cbdgef fgaecd cgeb fdcge agebfd fecdb fabcd edb | fdgacbe cefdb cefbgd gcbe\n', 
        'edbfga begcd cbg gc gcadebf fbgde acbgfd abcde gfcbed gfec | fcgedb cgb dgebacf gc\n', 
        'fgaebd cg bdaec gdafb agbcfd gdcbef bgcad gfac gcb cdgabef | cg cg fdcagb cbg\n',
        'fbegcd cbd adcefb dageb afcb bc aefdc ecdab fgdeca fcdbega | efabcd cedba gadfec cb\n',
        'aecbfdg fbg gf bafeg dbefa fcge gcbea fcaegb dgceab fcbdga | gecf egdcabf bgf bfgea\n',
        'fgeab ca afcebg bdacfeg cfaedg gcfdb baec bfadeg bafgc acf | gebdcfa ecba ca fadegcb\n',
        'dbcfg fgd bdegcaf fgec aegbdf ecdfab fbedc dacgb gdcebf gf | cefg dcbef fcge gbcadfe\n',
        'bdfegc cbegaf gecbf dfcage bdacg ed bedf ced adcbefg gebcd | ed bcgafe cdgba cbgef\n',
        'egadfb cdbfeg cegd fecab cgb gbdefca cg fgcdab egfdb bfceg | gbdfcae bgc cg cgb\n',
        'gcafb gcf dcaebfg ecagb gf abcdeg gaef cafbge fdbac fegbdc | fgae cfgab fg bagce\n'    
    ]
    
    signals_and_output_2 = [line.strip().split(" | ") for line in example_2]
    signals_2 = [item[0].split() for item in signals_and_output_2]
    output_2 = [item[1].split() for item in signals_and_output_2]
    output_value_2 = get_sum_of_output_values(signals_2, output_2)
    
    assert output_value_2 == 61229

    # Input file
    with open('data/input.txt') as f:
        data = f.readlines()
    signals_and_output = [line.strip().split(" | ") for line in data]
    signals = [item[0].split() for item in signals_and_output]
    output = [item[1].split() for item in signals_and_output]
    
    print(f"Sum of all output values: {get_sum_of_output_values(signals, output)}")
```
